```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Data augmentation & rescaling
train_gen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    zoom_range=0.15,
    horizontal_flip=True,
    vertical_flip=True,
    validation_split=0.2
)

# Training data
train_data = train_gen.flow_from_directory(
    train_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='training'
)

# Validation data
val_data = train_gen.flow_from_directory(
    train_path,
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(train_data, validation_data=val_data, epochs=10)  # Start with 10 epochs first
```

```python
import os
import shutil
import random

# Paths
data_dir = r'C:\Users\KIIT0001\Downloads\brain_tumour_ml_quantum\data'
train_dir = os.path.join(data_dir, 'train')
test_dir = os.path.join(data_dir, 'test')

# Create train/test folders if they don't exist
for folder in ['train/tumour', 'train/no_tumour', 'test/tumour', 'test/no_tumour']:
    os.makedirs(os.path.join(data_dir, folder), exist_ok=True)

# Original Kaggle folders
yes_dir = os.path.join(data_dir, 'yes')
no_dir = os.path.join(data_dir, 'no')

# Function to split images
def split_data(source_dir, train_subdir, test_subdir, split_ratio=0.8):
    images = [f for f in os.listdir(source_dir) if f.endswith(('.jpg', '.png'))]
    random.shuffle(images)
    split_index = int(len(images) * split_ratio)
    
    for i, img in enumerate(images):
        src_path = os.path.join(source_dir, img)
        if i < split_index:
            dst_path = os.path.join(train_subdir, img)
        else:
            dst_path = os.path.join(test_subdir, img)
        shutil.copyfile(src_path, dst_path)

# Split 'yes' images → tumour
split_data(yes_dir, os.path.join(data_dir, 'train/tumour'), os.path.join(data_dir, 'test/tumour'))

# Split 'no' images → no_tumour
split_data(no_dir, os.path.join(data_dir, 'train/no_tumour'), os.path.join(data_dir, 'test/no_tumour'))

print("Dataset split completed!")
```

```python
from qiskit import QuantumCircuit
from qiskit.providers.aer import AerSimulator
from qiskit import transpile, assemble


def cnn_to_quantum_confidence(prob):
    import numpy as np
    
    # Map probability to rotation angle
    theta = 2 * np.arcsin(np.sqrt(prob))
    
    # Single qubit circuit
    qc = QuantumCircuit(1,1)
    qc.ry(theta, 0)
    qc.measure(0,0)
    
    # Run on simulator
    backend = AerSimulator()
    t_qc = transpile(qc, backend)
    qobj = assemble(t_qc, shots=1024)
    result = backend.run(qobj).result()
    counts = result.get_counts()
    
    # Probability of measuring |1>
    quantum_confidence = counts.get('1',0)/1024
    return quantum_confidence

# Test it
q_conf = cnn_to_quantum_confidence(prob)
print(f"Quantum confidence of tumour: {q_conf:.2f}")

```

```python
# ---------------------------
# STEP 0: Imports
# ---------------------------
import os
import random
import shutil
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.models import load_model
from qiskit import QuantumCircuit, transpile, execute
from qiskit_ibm_provider import IBMProvider

# ---------------------------
# STEP 1: IBM Quantum Setup
# ---------------------------
# Replace 'YOUR_API_TOKEN' with your IBM Quantum API token
provider = IBMProvider(token='7ROMFLnWvd3oUv_wuJTbdZJk8EPsegf42UUqhzzu2Df7')
backend = provider.get_backend('ibmq_qasm_simulator')
print("IBM Quantum backend ready!")

# ---------------------------
# STEP 2: Dataset Paths
# ---------------------------
data_dir = r"C:\Users\KIIT0001\Downloads\brain_tumour_ml_quantum\data"
train_dir = os.path.join(data_dir, "train")
test_dir = os.path.join(data_dir, "test")

# Verify folders
for folder in ['tumour','no_tumour']:
    print("Train", folder, "images:", len(os.listdir(os.path.join(train_dir, folder))))
    print("Test", folder, "images:", len(os.listdir(os.path.join(test_dir, folder))))

# ---------------------------
# STEP 3: Load CNN model
# ---------------------------
model = load_model(r"C:\Users\KIIT0001\Downloads\brain_tumour_ml_quantum\models\cnn_brain.h5")
print("CNN model loaded!")

# ---------------------------
# STEP 4: Quantum confidence function
# ---------------------------
def cnn_to_quantum_confidence(prob):
    """
    Convert CNN probability to quantum confidence via IBM simulator.
    """
    theta = 2 * np.arcsin(np.sqrt(prob))
    
    qc = QuantumCircuit(1,1)
    qc.ry(theta,0)
    qc.measure(0,0)
    
    job = execute(transpile(qc, backend), backend=backend, shots=1024)
    result = job.result()
    counts = result.get_counts()
    
    q_conf = counts.get('1',0)/1024
    return q_conf

# ---------------------------
# STEP 5: Prediction on test set
# ---------------------------
test_images = []
cnn_probs = []
quantum_probs = []
labels = []

for label in ['tumour','no_tumour']:
    folder_path = os.path.join(test_dir, label)
    for img_file in os.listdir(folder_path):
        img_path = os.path.join(folder_path, img_file)
        img = load_img(img_path, target_size=(128,128))
        img_array = np.expand_dims(img_to_array(img)/255.0, axis=0)
        
        # CNN probability
        prob = model.predict(img_array)[0][0]  # assuming binary classification
        q_conf = cnn_to_quantum_confidence(prob)
        
        # Store
        test_images.append(img_file)
        cnn_probs.append(prob)
        quantum_probs.append(q_conf)
        labels.append(label)
        
        print(f"{img_file} | CNN: {prob:.2f} | Quantum: {q_conf:.2f} | Label: {label}")

# ---------------------------
# STEP 6: Visualization
# ---------------------------
plt.figure(figsize=(10,5))
plt.scatter(range(len(cnn_probs)), cnn_probs, color='blue', label='CNN Probability')
plt.scatter(range(len(quantum_probs)), quantum_probs, color='red', label='Quantum Confidence')
plt.xlabel("Test Images")
plt.ylabel("Probability / Confidence")
plt.title("CNN vs Quantum Confidence")
plt.legend()
plt.show()
```
