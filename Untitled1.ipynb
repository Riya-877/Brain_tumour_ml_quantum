```python
from qiskit import QuantumCircuit, Aer, execute
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
# Create a 2-qubit quantum circuit
qc = QuantumCircuit(2, 2)  # 2 qubits, 2 classical bits

# Apply Hadamard gate to qubit 0 (superposition)
qc.h(0)

# Apply CNOT gate (entanglement)
qc.cx(0, 1)

# Measure both qubits
qc.measure([0, 1], [0, 1])

# Draw the circuit
qc.draw('mpl')

```

```python
# Use the local qasm simulator
simulator = Aer.get_backend('qasm_simulator')

# Execute the circuit on the simulator
job = execute(qc, simulator, shots=1024)

# Get results
result = job.result()

# Get counts (measurement outcomes)
counts = result.get_counts(qc)
print("Measurement results:", counts)
```

```python
# Plot histogram of measurement results
plot_histogram(counts)
plt.show()
```

```python
# 3-qubit GHZ state example
qc2 = QuantumCircuit(3, 3)
qc2.h(0)
qc2.cx(0, 1)
qc2.cx(0, 2)
qc2.measure([0,1,2],[0,1,2])

job2 = execute(qc2, Aer.get_backend('qasm_simulator'), shots=1024)
result2 = job2.result()
counts2 = result2.get_counts()
print(counts2)

plot_histogram(counts2)
plt.show()
```
